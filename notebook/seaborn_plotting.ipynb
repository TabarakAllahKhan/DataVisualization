## Plotting Using Seaborn Library

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Important things for plotting according to grammer of graphics:

1. Data should be in the DataFrame format.
2. Type of plot?
3. X-Axis.
4. y Axis
5. Grouping (hue/color) based on categorical variables.
6. Size of plot.
7. Color pallets of the plot.
8. Style of the plot.
9. How do you interpret.

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

## Scatter Plot

```python
# scatter plot
# define size of plot
plt.figure(figsize=(12,8))
palette={'male':'green','female':'purple'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
# Labels

plt.xlabel('sex')
plt.ylabel('fare')
plt.title("Scatter Plot of Age VS Fare")
```

```python
# styling the plot

sns.set_theme(style='darkgrid',font_scale=1.5,font='Times New Roman')

# scatter plot
# define size of plot
plt.figure(figsize=(7,6))
palette={'male':'green','female':'purple'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',size='fare',palette=palette)
# Labels

plt.xlabel('Age')
plt.ylabel('fare')
plt.title("Scatter Plot of Age VS Fare")

```

## Interpretation

1. We have age of pessanger at X-axis and fare at y-axis.
2. The scatter plot shows us the relationship be the Age and fare of the titanic dataset.
3. We further grouped the passengers by their genders such as male in green points and purple for females.
4. Two Male passengers paid the highest fare.

## Box Plot

```python
# styling the plot

sns.set_theme(style='darkgrid',font_scale=1.5,font='Times New Roman')

# scatter plot
# define size of plot
plt.figure(figsize=(7,6))
palette={'male':'green','female':'purple'}
sns.boxplot(data=df,x='sex',y='fare',hue='sex',palette=palette)
# Labels

plt.xlabel('Age')
plt.ylabel('fare')
plt.title("Scatter Plot of Age VS Fare")

```

```python
# styling the plot

sns.set_theme(style='darkgrid',font_scale=1.5,font='Times New Roman')

# scatter plot
# define size of plot
plt.figure(figsize=(12,6))
palette={'First':'green','Second':'purple','Third':'red'}
sns.boxplot(data=df,x='sex',y='age',hue='class',palette=palette)
# Labels

plt.xlabel('SEX')
plt.ylabel('age')
plt.title("Box Plot of Age VS Sex")

```

## Boxen Plot

```python
# styling the plot

sns.set_theme(style='darkgrid',font_scale=1.5,font='Times New Roman')

# scatter plot
# define size of plot
plt.figure(figsize=(12,6))
palette={'First':'green','Second':'purple','Third':'red'}
sns.boxenplot(data=df,x='sex',y='age',hue='class',palette=palette)
# Labels

plt.xlabel('SEX')
plt.ylabel('age')
plt.title("Boxen Plot of Age VS Sex")
plt.xticks(ticks=[0,1],labels=['Male','Female'])
# rotate xticks by 45 degrees 

plt.xticks(rotation=45)

```
